```python
# Assignment - Recursion:

# Global Variable
WALL = '#'

from PIL import Image, ImageDraw

def solveFrom(maze, rowPos, colPos):
    
    if maze[rowPos][colPos] == 'F':
        return True
 
    if maze[rowPos][colPos] == WALL or maze[rowPos][colPos] == 'X':
        return False
  
    maze[rowPos][colPos] = 'X'

   
    if (solveFrom(maze, rowPos - 1, colPos) or  # up
        solveFrom(maze, rowPos + 1, colPos) or  # down
        solveFrom(maze, rowPos, colPos - 1) or  # left
        solveFrom(maze, rowPos, colPos + 1)):    # right
        return True

   
    maze[rowPos][colPos] = ' '
    return False

def save_maze_image(maze, output_filename):
    cell_size = 10
    height = len(maze)
    width = len(maze[0])

    img = Image.new('RGB', (width * cell_size, height * cell_size), (255, 255, 255))  # white
    draw = ImageDraw.Draw(img)

    for row in range(height):
        for col in range(width):
            ch = maze[row][col]
            color = 'black' if ch == WALL else 'white'
            if ch == 'X':
                color = 'green'
            # Draw the cell
            draw.rectangle([col * cell_size, row * cell_size, (col + 1) * cell_size, (row + 1) * cell_size], fill=color)

    img.save(output_filename)
    img.show()


maze = []
with open('maze_small.txt', 'r') as datafile:
    for line in datafile:
        rowList = [ch for ch in line if ch != '\n']
        maze.append(rowList)

# Print the list
for row in maze:
    print(row)


start_row, start_col = None, None
for y in range(len(maze)):
    for x in range(len(maze[y])):
        if maze[y][x] == 'S':
            start_row, start_col = y, x
            break
    if start_row is not None:
        break

print("Start Row(Y):", start_row, "Start Col(X):", start_col)

# Solve the maze
if solveFrom(maze, start_row, start_col):
    print('Path to finish found!')
else:
    print('No path to finish found.')

# Print the 2D list to confirm the structure
for row in maze:
    print(row)

save_maze_image(maze, 'solution.jpg')
```
